```python
import pandas as pd 
import requests 
from plyer import notification
from datetime import datetime

# Extraindo a tabela banks da API
url = "https://brasilapi.com.br/api/banks/v1"

bancos = requests.get(url)

data = datetime.now().strftime("%d/%m/%Y %H:%M:%S")

if bancos.status_code == 200:
    bancos_json = bancos.json()

# Transformando as tabelas em um DataFrame
bancos_full = pd.DataFrame(bancos_json)

# Alterando o nome das colunas
bancos_full_renomeado = bancos_full.set_axis(['ISPB', 'Nome', 'Código', 'Nome Completo'], axis=1)

# Ciando alerta de importação
def alerta():
    if bancos.status_code == 200:
        notification.notify(
            title = f'Atenção',
            message = f'Extração de dados realizada com sucesso em {data} da base {url}',
            timeout = 5
        )
    else:
        notification.notify(
            title = f'Atenção',
            message = f'Extração de dados não realizada, código do erro: {bancos.status_code} ',
            timeout = 5
        )

alerta()
bancos_full_renomeado
```
